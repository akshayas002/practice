```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.decomposition import PCA, KernelPCA
```

```python
X, y = make_moons(n_samples=200, noise=0.05, random_state=42)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.title("PCA Result (Fails)")
plt.show()
```

```python
kpca = KernelPCA(n_components=2, kernel='rbf', gamma=15)
X_kpca = kpca.fit_transform(X)
```

```python
plt.scatter(X_kpca[:, 0], X_kpca[:, 1], c=y, cmap='viridis')
plt.title("Kernel PCA Result (RBF Kernel)")
plt.show()
```

```python
kpca_1d = KernelPCA(n_components=1, kernel='rbf', gamma=15)
X_kpca_1d = kpca_1d.fit_transform(X)

plt.scatter(X_kpca_1d, np.zeros_like(X_kpca_1d), c=y, cmap='viridis')
plt.title("Kernel PCA (1D Projection)")
plt.show()
```
