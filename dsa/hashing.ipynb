```python
# counting frequencies of array elements
class Solution:
    def countFreq(arr, n):
        visited = [False]*n
        for i in range(n):
            if visited[i]:
              continue
            count = 1
            for j in range(i+1,n):
                if visited[i]==visited[j]:
                   visited[j]=True
                   count+=1
            print(arr[i], count)

if __name__ == "__main__":
    arr = [10, 5, 10, 15, 10, 5]
    n = len(arr)
    sol = Solution()
    sol.countFreq(arr, n)
```
