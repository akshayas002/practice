```python
# GCD of two numbers

# Brute force
def gcd(a,b):
    gcd=1
    for i in range(1,min(a,b)+1):
        if a%i==0 and b%i==0:
            gcd=i
    return gcd
print(gcd(18,12))

# Euclidian
def gcd(a,b):
    while b!=0:
        a, b = b, a%b
    return a
print(gcd(12,18))
```

```python
# Armstrong number
def is_armstrong(n):
    num=n
    k=len(str(n))
    sum = 0
    while n>0:
        ld = n%10
        sum = sum + ld**k
        n=n//10
    return sum == num
n = int(input())
print(is_armstrong(n))
```

```python
# Prime number
def is_prime(n):
    count=0
    for i in range(1,n//2):
        if n%i==0:
            count+=1
    return count == 2


n=int(input())
if is_prime(n)==True:
    print("not a prime number")
else:
    print("prime number")
```

```python
# Reverse integer

class Solution(object):
    def reverse(self, x):
        sign = -1 if x < 0 else 1
        x = abs(x)
        rev = 0
        while x != 0:
            digit = x % 10
            x = x // 10
            if rev > (2**31 - 1) // 10:
                return 0
            rev = rev*10 + digit
        return sign * rev
            
        
```

```python
# check if a number is a palindrome

class Solution(object):
    def isPalindrome(self, x):
        rev = 0
        og = x
        if x < 0:
            return False
        while x != 0:
            digit = x % 10
            x = x // 10
            rev = rev*10 + digit
        return rev==og
```
