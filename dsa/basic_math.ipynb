```python
# Count all digits of a number
n= int(input())
m=0
while n!=0:
    n=n//10
    m=m+1
print(m)
```

```python
# Reverse a number
n = input()
print(n[::-1])
```

```python
# palindrome number
n = input()
rev_n = n[::-1]
if rev_n==n:
    print('Palindrome number')
else:
    print("Not a palindrome number")
```

```python
# GCD of two numbers

# Brute force
def gcd(a,b):
    gcd=1
    for i in range(1,min(a,b)+1):
        if a%i==0 and b%i==0:
            gcd=i
    return gcd
print(gcd(18,12))

# Euclidian
def gcd(a,b):
    while b!=0:
        a, b = b, a%b
    return a
print(gcd(12,18))
```
