```python
# Remove Outermost Parentheses

def remPar(str):
    res = ''
    cnt = 0
    for c in str:
        if c=='(':
            cnt+=1
            if cnt>1:
                res+=c
        elif c==')':
            cnt-=1
            if cnt>0:
                res+=c
    return res

print(remPar("(()())(())(())"))
```

```python
# rotate array

def rotStr(str, goal):
    if len(str) != len(goal):
        return False
    for i in range(len(str)):
        rotated=str[i:]+str[:i]
        if rotated==goal:
            return True 
    return False
    

rotStr('rotation','tionrota')
```

```python
# Roman to integers

def romanToInt(s):
    roman = {
        'I':1, 'V':5, 'X':10, 'L':50, 'C':100, 'D':500, 'M':1000
    }
    res = 0
    for i in range(len(s)-1):
        if roman[s[i]]<roman[s[i+1]]:
            res-=roman[s[i]]
        else:
            res+=roman[s[i]]
    return res+roman[s[-1]]

print(romanToInt("MCMXCIV"))
```

```python
# sort characters by frequency
class Solution(object):
    def frequencySort(self, s):
        freq={}
        for c in s:
            freq[c] = freq.get(c,0)+1
            sorted_chars = sorted(freq.keys(), key=lambda x: (-freq[x], x))
        result = ""
        for ch in sorted_chars:
            result += ch * freq[ch]
        return result     
```

```python
# check if 2 strings are anagrams

class Solution(object):
    def isAnagram(self, s, t):
        if len(s)!=len(t):
            return False
        if sorted(s)==sorted(t):
            return True
        else:
            return False

        
```

```python
# isomorphic string

def isIsomorphic(self, s, t):
        return len(set(zip(s, t))) == len(set(s)) == len(set(t))
```

```python
# Count and Say

def countAndSay(self, n):
        result = "1"
        
        for _ in range(n - 1):
            new = ""
            i = 0
            
            while i < len(result):
                count = 1
                while i + 1 < len(result) and result[i] == result[i + 1]:
                    i += 1
                    count += 1
                
                new += str(count) + result[i]
                i += 1
            
            result = new
        
        return result
```

```python
# atoi

class Solution(object):
    def myAtoi(self, s):
        i = 0
        n = len(s)
        while i < n and s[i] == ' ':
            i += 1
        sign = 1
        if i < n and (s[i] == '+' or s[i] == '-'):
            if s[i] == '-':
                sign = -1
            i += 1
        num = 0
        INT_MAX = 2**31 - 1
        INT_MIN = -2**31
        while i < n and s[i].isdigit():
            digit = int(s[i])
            if num > (INT_MAX - digit) // 10:
                return INT_MAX if sign == 1 else INT_MIN

            num = num * 10 + digit
            i += 1

        return sign * num
       
```
