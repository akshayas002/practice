```python
# Remove Outermost Parentheses

def remPar(str):
    res = ''
    cnt = 0
    for c in str:
        if c=='(':
            cnt+=1
            if cnt>1:
                res+=c
        elif c==')':
            cnt-=1
            if cnt>0:
                res+=c
    return res

print(remPar("(()())(())(())"))
```

```python
# rotate array

def rotStr(str, goal):
    rotated=''
    for i in range(len(str)):
        rotated=str.substring(i)+str.substring(0,i)
        if rotated!=goal:
            continue
        else:
            return True
    

str = 'hello'
rotStr(str,'lehlo')
```
