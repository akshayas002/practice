```python
# Remove Outermost Parentheses

def remPar(str):
    res = ''
    cnt = 0
    for c in str:
        if c=='(':
            cnt+=1
            if cnt>1:
                res+=c
        elif c==')':
            cnt-=1
            if cnt>0:
                res+=c
    return res

print(remPar("(()())(())(())"))
```

```python
# rotate array

def rotStr(str, goal):
    if len(str) != len(goal):
        return False
    for i in range(len(str)):
        rotated=str[i:]+str[:i]
        if rotated==goal:
            return True 
    return False
    

rotStr('rotation','tionrota')
```

```python
def rotStrO(str, goal)->bool:
    return len(str) == len(goal) and goal in (str + str)
rotStrO('rotation','tionrota')
```

```python
# Roman to integers

def romanToInt(s):
    roman = {
        'I':1, 'V':5, 'X':10, 'L':50, 'C':100, 'D':500, 'M':1000
    }
    res = 0
    for i in range(len(s)-1):
        if roman[s[i]]<roman[s[i+1]]:
            res-=roman[s[i]]
        else:
            res+=roman[s[i]]
    return res+roman[s[-1]]

print(romanToInt("MCMXCIV"))
```

```python
# sort characters by frequency
class Solution(object):
    def frequencySort(self, s):
        freq={}
        for c in s:
            freq[c] = freq.get(c,0)+1
            sorted_chars = sorted(freq.keys(), key=lambda x: (-freq[x], x))
        result = ""
        for ch in sorted_chars:
            result += ch * freq[ch]
        return result
        
        
        
```
