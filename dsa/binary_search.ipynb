```python
class Solution(object):
    def search(self, nums, target):
        n = len(nums)
        low = 0
        high = n-1
        while low<=high:
            mid = (low + high) // 2
            if nums[mid]==target:
                return mid
            elif target>nums[mid]:
                low=mid+1
            else:
                high=mid-1
        return -1
```

```python
class Solution(object):

    def calTotalHrs(self, piles, speed):
        totalH = 0
        for bananas in piles:
            totalH += (bananas + speed - 1) // speed
        return totalH

    def minEatingSpeed(self, piles, h):
        low, high = 1, max(piles)
        ans = high
        
        while low <= high:
            mid = (low + high) // 2
            totalH = self.calTotalHrs(piles, mid)
            
            if totalH <= h:
                ans = mid
                high = mid - 1
            else:
                low = mid + 1
        
        return ans
```

```python
# Square root of a number

def sqrt(x):
        left, right , ans = 1, x//2, 0
        while left<=right:
            mid = (left+right)//2
            if mid*mid <= x:
                ans=mid
                left = mid +1
            else:
                right = mid - 1
        return ans

print(sqrt(1001))
        
```

```python
# BS on rotated array

class Solution(object):
    def search(self, nums, target):
        l, r= 0, len(nums)-1
        while l<=r:
            m = (l+r)//2
            if nums[m]==target:
                return m
            if nums[l]<=nums[m]:
                if nums[l]<=target<nums[m]:
                    r = m-1
                else:
                    l=m+1
            else:
                if nums[m]<target<=nums[r]:
                    l=m+1
                else:
                    r=m-1
        return -1
```
