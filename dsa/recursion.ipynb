```python
class Solution:
    def printName(self, name, count, N):
        if count == N:
            return
        print(name)
        self.printName(name, count + 1, N)
if __name__ == "__main__":
    sol = Solution()
    sol.printName('Ashish', 0, 10)
```

```python
class Solution:
    def printNum(self, current, n):
        if current > n:
            return
        print(current, end=' ')
        self.printNum(current+1, n)

if __name__ == '__main__':
    sol = Solution()
    n = 10
    sol.printNum(1,n)
```

```python
class Solution:
    def printNum(self, current, n):
        if current < n:
            return
        print(current, end=' ')
        self.printNum(current-1, n)

if __name__ == '__main__':
    sol = Solution()
    n = 10
    sol.printNum(n,1)
```

```python
class Solution:
    def sumNum(self, n):
        if n == 1:
            return 1
        return n+ self.sumNum(n-1)

if __name__=='__main__':
    sol = Solution()
    print(sol.sumNum(10))
```

```python
class Solution:
    def factorial(self, n):
        if n == 0:
            return 1
        return n*self.factorial(n-1)

if __name__=='__main__':
    sol = Solution()
    print(sol.factorial(4))
```

```python
class Solution:
    def revArr(self, arr):
        arr[:]=arr[::-1]

if __name__=='__main__':
    arr = [10, 20, 30, 40, 50]
    sol = Solution()
    sol.revArr(arr)
    print(" ".join(map(str, arr)))
```

```python
class Solution:
    def isPalindrome(self, i, s):
        if i >= len(s)//2:
            return True
        if s[i] != s[len(s)-1-i]:
            return False
        return(self.isPalindrome(i+1,s))
    
if __name__ == '__main__':
    s = input()
    sol = Solution()
    print(sol.isPalindrome(0,s))
```

```python
class Solution:
    def fibonacci(self, n):
        if n <= 1:
            return 1
        return self.fibonacci(n-1)+self.fibonacci(n-2)

if __name__ == '__main__':
    n = int(input())
    sol = Solution()
    print(sol.fibonacci(n))
```
